```python
import requests
response = requests.post("http://127.0.0.1:8000/token/", data={
            "username": 'admin',
            "password": 'user1234'
        })
response_json = response.json()
response_json
```

```python
import json
request_header = {"Content-Type": "application/json"}
request_body = {"refresh": response_json['refresh']}
response = requests.post("http://127.0.0.1:8000/token/refresh/",
                            data=json.dumps(request_body), headers=request_header)
response_json = response.json()
response_json
```

```python
request_header = {
            "Authorization": f"Bearer {response_json['access']}",
            "Content-Type": "application/json"
        }

response = requests.post("http://127.0.0.1:8000/usFunc2/test/", headers=request_header)
response.json()
```

```python
request_header = {
            "Authorization": f"Bearer {response_json['access']}",
            "Content-Type": "application/json"
        }
request_body = {"dataname":"close"}

response = requests.post("http://127.0.0.1:8000/usFunc2/quotes/", data=json.dumps(request_body), headers=request_header)
response.json()
```
